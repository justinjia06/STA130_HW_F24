### Research question 1: 
Is there a correlation between someone’s mental health and their views on being single?

**Variables:**
WELLNESS_satisfied_single
 - categorical (not satisfied, slightly not satisfied, satisfied, slightly satisfied, very satisfied)
 - Visualization: Bar plot to show how many people chose each level of satisfaction
WELLNESS_self_rated_mental_health
 - discrete (rate from 1(bad) - 10(great))
 - Visualization: Bar plot to show how many people chose each rating of mental health

**Analysis:**
Use fisher’s exact test on a 5x10 contingency table to calculate the exact probability of obtaining the observed distribution of values assuming the null hypothesis is true

**Hypothesis:**
Null hypothesis: There is no correlation between self-rated mental health and satisfaction with being single.
Alternative hypothesis: There is a correlation between self-rated mental health and satisfaction with being single.

**Results:**
If p-value < 0.05: reject the null hypothesis -> significant association 
If p-value >= 0.05: fail to reject the null hypothesis -> no significant association 
The results will help determine if there is a relationship between attitudes toward being single and their mental health

### Research question 2:
Does weight affect your happiness?

**Variables:**
WELLNESS_subjective_happiness_scale_happy
 - Discrete (rate from 1(sad) - 10(happy))
 - Visualization: Bar plot to show how many people chose each level of happiness
DEMO_weight
 - Continuous ( 0(probably not) - infinity(definitely not))
 - Visualization: violin plot to see the distribution shape and density of the data

**Analysis:**
Simple linear regression to analyze the potential effect of weight on a person's happiness score. Where happiness=β0+β1×Weight+ϵ

**Hypothesis:**
Null hypothesis: There is no relationship between weight and happiness, slope of the regression line is zero
Alternative hypothesis: There is a relationship between weight and happiness, slope of the regression line is not zero

**Results:**
If p < 0.05: We reject the null hypothesis -> statistically significant relationship 
 - If β1>0 -> as weight increases, happiness increases.
 - If β1<0 -> as weight increases, happiness decreases.
If p >= 0.05: We failed to reject the null hypothesis -> no statistically significant relationship
The results will help determine if there is a relationship between weight and self percieved happiness
